# Project 2: AMES Housing Sale Prices Prediction

## Problem Statement

Predicting the sale prices of residential housing in AMES, Iowa using advanced regression techniques and machine learning algorithms. 

## Executive Summary

This project investigates factors that influence decision-making calculus of potential home buyers and how various attributes of a property impacts its sales price. The aim of the project is to build a housing sale price prediction model using the most important attributes (or features) of the house. 

To achieve the objective of predicting sales prices of residential housing in Ames, we were provided with two different data sets: train and test. The Train Dataset comprised of observations of 2051 different properties (rows) and 81 different characteristics or features (columns) of the property. The columns comprised of 22 nominal, 21 ordinal and 38 continuous (numerical) continuous
variables. The two other columns were the identification columns. On the other hand, the Training data set provided the context to build a model for predicting sale prices. The test data set, for which sale prices were to be predicted, comprised of 879 observations (properties) and 80 features or characteristics. The test data set, unlike the training data set, does not include sale price as a variable since this was the value to be predicted.

### Filling up NaN values for categorical variables (different categories NA, zero,etc.)

```python
#filling up NaN values in categorical variables with None which means No feature present (or absence of a feature)

variables_1 = ['alley','garage_type','misc_feature','bsmt_qual','bsmt_cond','bsmt_exposure','bsmtfin_type_1',
               'bsmtfin_type_2','fireplace_qu','garage_finish','garage_qual','garage_cond','pool_qc',
               'fence','mas_vnr_type']

housing_train[variables_1]=housing_train[variables_1].fillna('None')
```

```python
#filling up zero for missing values in some of the numerical columns 
#mas_vnr_area is filled with 0 because for all those values mas_vnr_type is None => no vnr type present
#garage_yr_built is zero because for those values garage type (except one value where type is Detached but no other info is availale, thus it is ignored and all set to zero) and garage finish are NA which means no garage 
#bsmtfin_sf_1, sf_2,unf_sf,total_sf are zero because bsmt qual and bsmt fin type 1 and 2 are NA => no basement 
#bsmt_half_bath and bsmt_full_bath are zero because bsmt_cond set to NA => no basement
#garage_cars and garage_area are zero because garage_yr_blt is zero 

variables_2 = ['mas_vnr_area','garage_yr_blt','garage_cars','garage_area','bsmtfin_sf_1',
               'bsmtfin_sf_2','bsmt_unf_sf','total_bsmt_sf','bsmt_half_bath','bsmt_full_bath']

housing_train[variables_2]=housing_train[variables_2].fillna(int(0))
```

```python
#imputing the values for lot_frontage - Since it is hypothesized that households in the same neighborhood will have a similar sale price, that's why missing values are imputed for each neighborhood by its median

housing_train["lot_frontage"] = housing_train.groupby('neighborhood')["lot_frontage"].transform(lambda x: x.fillna(x.median()))
housing_train["lot_frontage"] = housing_train["lot_frontage"].fillna(housing_train["lot_frontage"].mean())
```

### Train-test split 

```python
#selecting X and y values
features = [col for col in housing_train._get_numeric_data().columns if col != 'saleprice'] 
X = housing_train[features]
y = housing_train['saleprice']
```

```python
#accessing only the numeric data in housing_train dataset
housing_train._get_numeric_data().columns
```

```python
#checking the dimension of X and y
X.shape
```

```python
y.shape
```

```python
#splitting the data into train and test set
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=50)
```

```python
#Scaling: fitting a standard scaler to X_train and using it to transform X_train and X_test
ss = StandardScaler()
ss.fit(X_train)
X_train_scaled = ss.transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
#checking the dimension of scaled X variables in train and test datasets
X_train_scaled.shape, X_test_scaled.shape
```

## Regression

### Linear Regression

```python
#fitting the scaled X_train and y_train with linear regression
lr = LinearRegression()
lr.fit(X_train_scaled,y_train)
```

```python
#plotting the lr coefficients
plt.plot(lr.coef_);
```

```python
#using cross-validation to evaluate the model
lr_scores = cross_val_score(lr,X_train_scaled,y_train,cv=10)
lr_scores.mean()
```

### Lasso Regression

```python
#fitting the scaled X_train and y_train with lasso regression
lasso = LassoCV(alphas=np.arange(0.01, 0.5, 0.02))
lasso.fit(X_train_scaled,y_train)
```

```python
#finding optimal lasso alpha value
lasso.alpha_
```

```python
#finding the mean R2 in lasso
lasso_scores = cross_val_score(lasso,X_train_scaled,y_train,cv=10)
lasso_scores.mean()
```

```python
#train RMSE
np.sqrt(mean_squared_error(y_train,lasso.predict(X_train_scaled)))
```

```python
#test RMSE
np.sqrt(mean_squared_error(y_test,lasso.predict(X_test_scaled)))
```

### Ridge Regression

```python
#fitting the scaled X_train and y_train with ridge regression
ridge = RidgeCV(alphas=np.linspace(0.1, 10, 100))
```

```python
#finding the mean R2 in ridge
ridge_scores = cross_val_score(ridge,X_train_scaled,y_train,scoring='r2',cv=10)
ridge_scores.mean()
```

### ElasticNet Regression 

```python
#fitting the scaled X_train and y_train with ElasticNet regression
ENet = ElasticNetCV(alphas=np.linspace(0.1, 10, 100))
ENet.fit(X_train_scaled,y_train)
```

```python
#finding the mean R2 in ridge
ENet_scores = cross_val_score(ENet,X_train_scaled,y_train,scoring='r2',cv=10)
ENet_scores.mean()
```

```python
#train RMSE
np.sqrt(mean_squared_error(y_train,ENet.predict(X_train_scaled)))
```

```python
#test RMSE
np.sqrt(mean_squared_error(y_test,ENet.predict(X_test_scaled)))
```

## Conclusion and Recommendations

I ran correlation on my training data set to find out which variables are strongly correlated with the dependent variable, the sale price. Variables such as overall quality, ground living area, garage area, garage cars, total basement area were found to be highly correlated with sales price. The correlation was also strong between some of the independent variables. For example, garage
cars and garage area were significantly correlated. To eliminate multi-collinearity, one of either variables had to be dropped while building the prediction model.

After dividing the training set into Train and Validation set, several models were fitted on the Train and tested on the Validation set to find the best model. The best model is the one which is able to predict the sale price most efficiently and precisely. I ran Linear Regression, Lasso Regression, Ridge Regression and ElasticNet Regression to find the highest accuracy score (i.e. R-squared value). The ElasticNet regression provided the best R-squared value and the Root Mean Squared error for the validation set slightly improved over the Root Mean Squared error of the Train set. Therefore, Elastic Net Regression is preferred over other methods and is applied to the validation test. ENet coefficients were calculated and the final prediction model was built using 22 variables including both categorical and continuous variables. Eventually, this prediction model was tested on the Housing Test Data set and sale price predictions were made. 

Feature engineering of some of the variables by using higher-order polynomial function is recommended to deal with the issue of overfitting of variables and selecting the most important variables. Some of the features can be combined with others to produce one feature such as total bathrooms could be calculated by weighting and summing up the half and full bathrooms. This approach would automatically eliminate some amount of multi-collinearity among the explanatory variables.
